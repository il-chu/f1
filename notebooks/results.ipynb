```python
import pandas as pd
import numpy as np
import os
```

```python

session = pd.read_csv(f"data/raw/2025/race_sesions.csv")
session = session.drop(columns=["circuit_key", "country_key", "gmt_offset", "date_end", "session_type", "session_name"])
meetings = pd.read_csv(f"data/raw/2025/meetings.csv")
meetings = meetings.drop(columns=["location", "country_key", "country_code", "country_name", "circuit_key", "circuit_short_name", "gmt_offset", "date_start", "date_end", "year" ])

strategy = pd.read_csv(f"data/processed/2025/strategic_metrics.csv")
dynamics = pd.read_csv(f"data/processed/2025/race_dynamics_metrics.csv")
chaos = pd.read_csv(f"data/processed/2025/chaos_metrics.csv")
championship_impact = pd.read_csv(f"data/processed/2025/championship_impact.csv")




df = meetings.merge(session, on="meeting_key", how="inner")
df = df.merge(strategy, on="session_key", how="inner")
df = df.merge(dynamics, on="session_key", how="inner")
df = df.merge(chaos, on="session_key", how="inner")
df = df.merge(championship_impact, on="session_key", how="inner")


df["meeting_name_short"] = df["meeting_name"].str.replace("Grand Prix", "GP", regex=False)
emoji_maping = {
    "italy": "it",
    "united states": "us",
    "japan": "jp", 
    "saudi arabia": "saudi_arabia",
    "china": "cn", 
    "spain": "es",
    "united kingdom": "uk",
    "united arab emirates": "united_arab_emirates"


}
df["flag_emoji"] = df["country_name"].str.lower().replace(emoji_maping, regex=False)

def minmax_scale(series):
    return (series - series.min()) / (series.max() - series.min())

df["strategy_score_scaled"] = minmax_scale(df["strategy_score"])
df["chaos_score_scaled"] = minmax_scale(df["chaos_score"])
df["championship_impact_score_scaled"] = minmax_scale(df["championship_impact_score"])
df["race_dynamics_score_scaled"] = minmax_scale(df["race_dynamics_score"])

df["race_score"] = 0.33*df["chaos_score_scaled"] + 0.33*df["strategy_score_scaled"] + 0.33*df["race_dynamics_score_scaled"]

df.to_csv("data/processed/2025/all_data.csv", index=False)
df
```
